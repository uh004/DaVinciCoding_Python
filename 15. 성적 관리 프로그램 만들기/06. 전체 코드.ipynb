```python
# 전체 코드
def print_menu():
    print("레오의 성적 관리 프로그램")
    print("무엇을 할까요?")
    print("1. 과목별 점수 보기")
    print("2. 전체 점수 보기")
    print("3. 점수 추가")
    print("4. 점수 삭제")
    print("5. 점수 삭제")
    print("6. 총점 보기")
    print("7. 평균 보기")
    print("0. 프로그램 종료")

def print_sub_score():
    subject = input("보고싶은 과목을 입력하세요")
    if subject not in scores.keys():
        print(f"{subject}은(는) 없는 과목 입니다")
        return

    print(f'{subject} 점수 = {scores[subject]}')

def print_all_scores():
    for k, v in scores.items():
        print(f'{k} 점수 = {v}')

def add_score():
    subject = input("추가할 과목을 입력하세요")
    if subject in scores.keys():
        print(f"{subject}는 이미 있는 과목 입니다")
        return

    score = int(input("추가할 점수를 입력하세요"))
    scores[subject] = score
    print("점수가 추가 되었습니다.")

def change_score():
    subject = input("수정할 과목을 입력하세요")
    if subject not in scores.keys():
        print(f"{subject}은(는) 없는 과목 입니다")
        return

    score = int(input("수정할 점수를 입력하세요"))
    scores[subject] = score
    print("점수가 수정 되었습니다.")

def del_score():
    subject = input("삭제할 과목을 입력하세요")
    if subject not in scores.keys():
        print(f"{subject}는 없는 과목 입니다")
        return
    del scores[subject]
    print("점수가 삭제 되었습니다.")

def get_total():
    total = 0
    for score in scores.values():
        total += score

    return total

scores = {'국어' : 80, '영어' : 75, '수학' : 90, '과학' : 85}

while True:
    print_menu()
    num = int(input())
    if num == 1: # 1. 과목별 점수 보기
        print_sub_score()
    elif num == 2:
        print_all_scores()
    elif num == 3: # 3. 점수 추가
        add_score()
    elif num == 4: # 4. 점수 수정
        change_score()
    elif num == 5: # 5. 점수 삭제
        del_score()
    elif num == 6: # 6. 총점 보기
        print(f'총점 = {get_total()}')
    elif num == 7: # 7. 평균 보기
        print(f'평균 = {sum(scores.values()) / len(scores)}')
    elif num == 0:
        print("프로그램을 종료합니다.")
        break
```
